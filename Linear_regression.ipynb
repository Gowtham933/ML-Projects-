#Linear Regression

It is a method used to traget the model relationship beteen a dependent varaiable(target) and one or more idenependent variebles(predoctors).

```python
# Sample Data
X = np.array([600, 800, 1000, 1200, 1400]).reshape(-1, 1)
y = np.array([180, 220, 260, 300, 340])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {'House Size(Sq.ft)': X.ravel(), 'Price($)': y} # Change X to X.ravel() to make it 1-dimensional
df = pd.DataFrame(data)
df
```

AIM:  Here we have house size and its corresponding price in $.

I need to fine the price for 1128 Sq.ft? which is not exist in the table.

Formula for Simple linear regression

Y=mX+b

where:

  -YY = dependent variable (output)

  -XX = independent variable (input)

  -mm = slope of the line (how much YY changes for each unit of XX)

  -bb = intercept (value of YY when X=0X=0)

```python
# Created and train the model
model = LinearRegression()
model.fit(X, y)

# Make a prediction?
predicted_price = model.predict([[1128]])
print(f"Predicted price for 1100 sq. ft house: ${predicted_price[0] * 1000}")

# Plot the data
plt.scatter(X, y, color='blue', label='Actual Prices')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel("House Size (sq. ft)")
plt.ylabel("Price (in $1000s)")
plt.legend()
plt.show()
```

**Result**: Predicted price for 1100 sq. ft house: $285600.0


Future process:

Here I demonstarte to predict multiple data point at the same time

```python
# Make a prediction - Pass a 2D array with one feature for each prediction
predicted_price = model.predict([[1128], [1134], [1257]])
# Print predictions for each house size
for size, price in zip([1128, 1134, 1257], predicted_price):
    print(f"Predicted price for {size} sq. ft house: ${price}")

# Plot the data
plt.scatter(X, y, color='blue', label='Actual Prices')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.xlabel("House Size (sq. ft)")
plt.ylabel("Price (in $1000s)")
plt.legend()
plt.show()
```

```python
# New data points for prediction
new_X = np.array([1128, 1134, 1257]).reshape(-1, 1)
predicted_price = model.predict(new_X)

# Create a DataFrame for the predicted values
predicted_df = pd.DataFrame({'House Size(Sq.ft)': new_X.ravel(), 'Predicted Price($)': predicted_price})
predicted_df

# Merge the original DataFrame with the predicted DataFrame
merged_df = pd.concat([df, predicted_df], ignore_index=True)
print(merged_df)

#Plot the data
plt.scatter(X, y, color='blue', label='Actual Prices')
plt.plot(X, model.predict(X), color='red', label='Regression Line')
plt.scatter(new_X, predicted_price, color='green', label='Predicted Prices') # Plot predicted points
plt.xlabel("House Size (sq. ft)")
plt.ylabel("Price (in $1000s)")
plt.legend()
plt.show()
```
